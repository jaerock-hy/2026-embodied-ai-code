# Logic Gates using Two-Input Perceptron

```python
def and_gate(a, b):
    w1, w2, th = 0.09, 0.05, 0.1
    return 1 if w1 * a + w2 * b >= th else 0

def nand_gate(a, b): 
    w1, w2, th = -0.09, -0.05, -0.1 # this isn't right.
    return 0 if w1 * a + w2 * b >= th else 1

def or_gate(a, b):
    w1, w2, th = 0.09, 0.05, 0.05
    return 1 if w1 * a + w2 * b >= th else 0

def nor_gate(a, b):
    w1, w2, th = -0.09, -0.05, -0.05 # this isn't right.
    return 0 if w1 * a + w2 * b >= th else 1
```

## LogicGate class

```python
class LogicGatePrimitive:
    def __init__(self):
        pass

    def and_gate(self, a, b):
        w1, w2, th = 0.09, 0.05, 0.1  
        return 1 if w1 * a + w2 * b >= th else 0
    
    def nand_gate(self, a, b):
        w1, w2, th = -0.5, -0.5, -0.7   # this isn't right.
        return 0 if w1 * a + w2 * b >= th else 1

    def or_gate(self, a, b):
        w1, w2, th = 0.09, 0.05, 0.05 
        return 1 if w1 * a + w2 * b >= th else 0

    def nor_gate(self, a, b):
        w1, w2, th = -0.09, -0.05, -0.1  # this isn't right.
        return 0 if w1 * a + w2 * b >= th else 1
```

```python
logic_gate = LogicGatePrimitive()
```

```python
logic_gate.and_gate(0, 0), logic_gate.and_gate(0, 1), logic_gate.and_gate(1, 0), logic_gate.and_gate(1, 1)
```

```python
logic_gate.nand_gate(0, 0), logic_gate.nand_gate(0, 1), logic_gate.nand_gate(1, 0), logic_gate.nand_gate(1, 1)
```

```python
logic_gate.or_gate(0, 0), logic_gate.or_gate(0, 1), logic_gate.or_gate(1, 0), logic_gate.or_gate(1, 1)
```
